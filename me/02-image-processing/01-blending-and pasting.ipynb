```python
import cv2 as cv
```

```python
img1 = cv.imread('../../course/DATA/dog_backpack.png')
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)

img2 = cv.imread('../../course/DATA/watermark_no_copy.png')
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
```

```python
img1_1200x1200 = cv.resize(img1, (1200,1200))
img2_1200x1200 = cv.resize(img2, (1200,1200))
```

```python
large_img = img1
small_img = cv.resize(img2, (600,600))
```

```python
x_offset = round((large_img.shape[1] - small_img.shape[1])/2)
y_offset = round((large_img.shape[0] - small_img.shape[0])/2)
print(x_offset)
print(y_offset)
```

```python
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

```python
new_img = large_img.copy()
new_img[y_offset : y_end, x_offset : x_end] = small_img
```

```python
plt.imshow(new_img)
```
