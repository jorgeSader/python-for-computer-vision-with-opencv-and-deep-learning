```python
img = cv.imread('../../course/DATA/crossword.jpg',0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
def show_pic(img):
  fig = plt.figure(figsize=(15,15))
  ax = fig.add_subplot(111)
  ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
ret, th1 = cv.threshold(img,180, 255, cv.THRESH_BINARY)
ret
```

```python
show_pic(th1)
```

```python
th2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 16)
show_pic(th2)
```

```python
blended = cv.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=0.4, gamma=0)
show_pic(blended)
```
