```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2 as cv
print(cv.__version__)
```

```python
img=cv.imread('../course/DATA/00-puppy.jpg')
```

```python
#MATPLOTLIB --> RGB (RED GREEN BLUE)
#OPENCV --> BGR (BLUE GREEN RED)

#we need to convert using cv.cvtColor(img, conversion_mode)
fix_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv.imread('../course/DATA/00-puppy.jpg', cv.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
plt.imshow(fix_img)
```

```python
new_img = cv.resize(fix_img, (1000, 400))
```

```python
plt.imshow(new_img)
```

```python
w_ratio = 1
h_ratio = 0.3
```

```python
half_img = cv.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
```

```python
plt.imshow(half_img)
```

```python
img_fliped_x = cv.flip(fix_img, 0)
plt.imshow(img_fliped_x)
```

```python
img_fliped_y = cv.flip(fix_img, 1)
plt.imshow(img_fliped_y)
```
