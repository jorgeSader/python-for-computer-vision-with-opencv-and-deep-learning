```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
pic = Image.open('../../course/DATA/00-puppy.jpg')
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
```

```python
type(pic_arr)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
```

```python
plt.imshow(pic_red)
```

```python
# Python uses RGB (Red, Green, Blue) which means that if we remember the shape of pic_arr

pic_arr.shape
```

```python
# red is the first plane array and we can isolate it by selecting every row, Every column and just the first plane of the shape (at index 0).
plt.imshow(pic_red[:,:,0])
```

```python
# Green is the second plane array and we can isolate it by selecting every row, Every column and just the second plane of the shape (at index 1).
pic_green = pic_arr.copy()
plt.imshow(pic_green[:,:,1])
```

```python
# Blue is the third plane array and we can isolate it by selecting every row, Every column and just the third plane of the shape (at index 1).
pic_blue = pic_arr.copy()
plt.imshow(pic_blue[:,:,2])
```

```python
# As we can tell, they are all a bit diferent as, while they are all reading values from 0 to 255, the values are based on a diferent diferent color. 
# To make this somewhat easier to understand, we often look at the grayscale version of such images by setting a seccond argument, cmap, to 'gray'.
```

```python
plt.imshow(pic_red[:,:,0], cmap = 'gray')
```

```python
plt.imshow(pic_green[:,:,1], cmap = 'gray')
```

```python
plt.imshow(pic_blue[:,:,2], cmap = 'gray')
```

```python
# We could also see the blue values only by setting all Red and Green values to 0.

pic_blue = pic_arr.copy()
pic_blue[:,:,0]=0
pic_blue[:,:,1]=0
```

```python
plt.imshow(pic_blue)
```

```python
# or just the green

pic_green = pic_arr.copy()
pic_green[:,:,0]=0
pic_green[:,:,2]=0

plt.imshow(pic_green)
```

```python
# or red

pic_red = pic_arr.copy()
pic_red[:,:,1]=0
pic_red[:,:,2]=0

plt.imshow(pic_red)
```
